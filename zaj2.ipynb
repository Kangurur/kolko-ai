# Klasy w Pythonie

```python
class Person:
    species = 'człowiek'   #zmienna statyczna

    def __init__(self, name, age):
        self.name = name      #zmienne klasowe
        self.age = age

    def greet(self):
        return f"Cześć, jestem {self.name} i mam {self.age} lat."

    def __repr__(self):
        return f"Person(name={self.name!r}, age={self.age})"
    
    def rev_name(self):
        return self.name[::-1]
    
    def x_name(self,x):
        return x*self.name
```

```python
Bob = Person("Bob", 18)
print(Bob.greet())
print(Bob.rev_name())
print(Bob.x_name(3))
print(Bob)
print(Bob.species)
```

Dziedziczenie

```python
class Uczeń(Person):
    def __init__(self, name, age, school):
        super().__init__(name, age)
        self.school = school

    def greet(self):
        return super().greet() + f" Uczę się w {self.school}."


```

```python
s = Uczeń('Marek', 17, 'III LO')
print(s.greet())
print(s.rev_name())
```


- Metody instancji muszą przyjmować `self` jako pierwszy parametr.
- `__init__` inicjalizuje obiekt.
- `__repr__` daje czytelną reprezentację obiektu.

## Matplotlib — wykresy


```python
import matplotlib.pyplot as plt
import numpy as np

# 1) wykres liniowy
x = np.linspace(-2*np.pi, 2*np.pi, 200)
y1 = np.sin(x)
y2 = np.cos(x)
plt.figure(figsize=(6,3))
plt.plot(x, y1, label='sin(x)')
plt.plot(x, y2, label='cos(x)', linestyle='--')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Wykres liniowy: sin i cos')
plt.legend()
plt.grid()
plt.show()

# 2) histogram
data = np.random.default_rng(42).normal(loc=5, scale=2, size=200)
plt.figure(figsize=(5,3))
plt.hist(data, bins=30, color='cadetblue', edgecolor='black')
plt.title('Histogram')
plt.tight_layout()
plt.show()

# 3) imshow (macierz jako obraz)
M = np.arange(36).reshape(6,6)
plt.figure(figsize=(4,3))
plt.imshow(M, cmap='plasma')
plt.colorbar()
plt.title('Imshow: macierz 6x6')
plt.tight_layout()
plt.show()

# 4) wykres liniowy 2

def f(x):
    return (np.sin(3*x) * np.exp(np.sin(x**2)) + 0.3 * np.sin(80*x) / (1 + 0.5*x**2) + np.cos(0.5 * x**3) * 0.5)
    
x = np.linspace(-5, 5, 1000)
y = f(x)
plt.figure(figsize=(6,3))
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

## Transformers

```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer

model_name = "gpt2"
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = GPT2LMHeadModel.from_pretrained(model_name)

prompt = "i like dogs because"

inputs = tokenizer(prompt, return_tensors="pt")

outputs = model.generate(
    **inputs,
    max_new_tokens=20,
    temperature=0.7,
    top_p=0.9,
    do_sample=True,
)

print(tokenizer.decode(outputs[0], skip_special_tokens=True))
```

```python
# to samo
from transformers import pipeline

model_name = "gpt2"
gen = pipeline("text-generation", model=model_name, tokenizer=model_name)

prompt = "i like dogs because"
out = gen(prompt, max_new_tokens=20, do_sample=True, temperature=0.7, top_p=0.9)

print(out[0]["generated_text"])
```

```python
from transformers import pipeline

model_name = "google/flan-t5-small"
# używamy text2text-generation (T5/FLAN są modelami text2text)
gen = pipeline("text2text-generation", model=model_name, tokenizer=model_name)

prompt2 = "Translate Polish to English: kocham kolko ai"
out2 = gen(prompt2, max_new_tokens=40)
print('\n', out2[0]["generated_text"])
```
